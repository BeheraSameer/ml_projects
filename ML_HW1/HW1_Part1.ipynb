```python
import numpy as np
```

```python
xtrn = np.array([0,2,3,5])
ytrn = np.array([1,4,9,16])
xtst = np.array([1,4])
ytst = np.array([3,12])
xtrace = np.linspace(0,5,20)
```

```python
from scipy.interpolate import *
```

```python
yreg = []
for d in range(4):
    yreg_calc = np.polyfit(xtrn,ytrn,d)
    yreg.append(yreg_calc)
```

```python
import  matplotlib.pyplot as plt
```

```python
plt.figure(1)
plt.plot(xtrn,ytrn,'X',label='Training Data')
plt.plot(xtst,ytst,'o',label='Testing Data')
for d in range(4):
    plt.plot(xtrace,np.polyval(yreg[d],xtrace),label='Degree-{}'.format(d))
plt.title("Polynomial Regression Plots")
plt.legend(loc='best')
plt.xlabel('X')
plt.ylabel('Y')
```

```python
total_error_array = []
training_error_array = []
for d in range(4):
    train_rse_error = 0
    for i in range(len(xtrn)):
        train_rse_error += pow((np.polyval(yreg[d],xtrn[i])-ytrn[i]),2)
    total_error_array.append(train_rse_error)
    print("Training Error Data for Degree {} is".format(d),train_rse_error)
    training_error_array.append(train_rse_error)
print("\n")

test_error_array = []
for d in range(4):
    test_rse_error = 0
    for i in range(len(xtst)):
        test_rse_error += pow((np.polyval(yreg[d],xtst[i])-ytst[i]),2)
    total_error_array[d]+=test_rse_error
    print("Test Error Data for Degree {} is".format(d),test_rse_error)
    test_error_array.append(test_rse_error)
print("\n")

for d in range(4):
    print("Total Error Data for Degree {} is".format(d),total_error_array[d])
```

```python
plt.figure(2)
plt.plot(range(4),training_error_array,label='Training Error')
plt.plot(range(4),test_error_array,label='Test Error')
plt.plot(range(4),total_error_array,'m-.',label='Total Error')
plt.title("Error Plots")
plt.legend(loc='best')
plt.xlabel('Degree')
plt.ylabel('Error Value')
```
